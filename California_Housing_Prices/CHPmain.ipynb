```python
df = df.copy()
missing_bedrooms = df['total_bedrooms'].isnull()
df_with_bedrooms = df[~missing_bedrooms]
df_missing_bedrooms = df[missing_bedrooms]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(df_with_bedrooms[['total_rooms', 'households']], df_with_bedrooms['total_bedrooms'])
```

```python
pred_br = lm.predict(df_missing_bedrooms[['total_rooms', 'households']])
```

```python
df.loc[missing_bedrooms, 'total_bedrooms'] = pred_br
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('median_house_value', axis=1)
y = df['median_house_value']
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
numerical_cols = df.drop(['ocean_proximity', 'median_house_value'], axis=1).columns

preprocessor = ColumnTransformer(
    transformers = [
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), ['ocean_proximity'])
    ])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor

model_rf = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', RandomForestRegressor(random_state=42))
])
```

```python
param_grid = {
    'regressor__n_estimators': [10, 30, 100],
    'regressor__max_features': [2, 4, 6, 8],
    'regressor__bootstrap': [False, True]
}

grid_search = GridSearchCV(model_rf, param_grid, cv=5, scoring='neg_mean_squared_error', return_train_score=True)
grid_search.fit(X_train, y_train)

print(grid_search.best_params_)
```

```python
from sklearn.metrics import mean_squared_error

best_model_rf = grid_search.best_estimator_
y_pred_rf_tuned = best_model_rf.predict(X_test)

print("R^2: {}".format(best_model_rf.score(X_test, y_test)))

rmse_rf_tuned = np.sqrt(mean_squared_error(y_test, y_pred_rf_tuned))
print("Root Mean Squared Error: {}".format(rmse_rf_tuned))
```

```python
sample_data = X_test.iloc[:10]

predicted_prices = best_model_rf.predict(sample_data)

actual_prices = y_test.iloc[:10].values

for actual, predicted in zip(actual_prices, predicted_prices):
    print(f"Actual Price: ${actual:.2f} | Predicted Price: ${predicted:.2f}")
```
