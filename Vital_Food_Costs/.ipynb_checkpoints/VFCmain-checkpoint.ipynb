```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Food Prices.csv')
```

```python
df.columns = df.columns.str.strip()
```

```python
# Creating a 'Date' column combining 'Year' and 'Month' for easier time-series analysis
df['Date'] = pd.to_datetime(df[['Year', 'Month']].assign(DAY=1))
```

```python
# Extracting the year from the 'Date' column and overwriting the 'Year' column for consistency
df['Year'] = df['Date'].dt.year
```

```python
annual_price_increase = df.groupby(['Country', 'Food Item', 'Year'])['Price in USD'].mean().pct_change()
annual_price_increase_df = annual_price_increase.reset_index()
display(annual_price_increase_df.style.format({
    'Price in USD': '{:,.2f}'.format,
}).set_caption('Annual price increase'))
```

#### There was a significant surge in egg prices in Australia between 2020 and 2022.

```python
avg_inflation = annual_price_increase.groupby(['Country', 'Food Item']).mean()
avg_inflation_df = avg_inflation.reset_index()
display(avg_inflation_df.style.format({
    0: '{:,.2%}'.format,
}).set_caption('Average inflation'))
```

```python
# This analysis provides a simplified representation of the average price of these essential food items in each country. 
df_country_year = df.groupby(['Country', 'Year'])['Price in USD']
avg_price = df_country_year.mean().unstack()
display(avg_price.style.format('{:,.2f}'.format).set_caption('Average price per year'))
```

```python
annual_avg_price = df_country_year.mean().reset_index()
display(annual_avg_price.style.format({
    'Price in USD': '{:,.2f}'.format,
}).set_caption('Annual average price'))
```

```python
plt.figure(figsize=(15,10))
sns.set_style('whitegrid')
sns.lineplot(data=df, x='Date', y='Price in USD', hue='Country', style='Food Item', palette='tab10', linewidth=2.5)
plt.title('Trends in Prices Over Time for Each Country and Food Item', fontsize=18)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Price in USD', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

#### Food prices in South Africa are relatively cheaper compared to other four countries




```python
plt.figure(figsize=(20,15))
sns.set_style('whitegrid')
sns.boxplot(data=df, x='Food Item', y='Price in USD', hue='Country', palette='tab10')
plt.title('Distribution of Prices for Each Food Item in Each Country', fontsize=18)
plt.xlabel('Food Item', fontsize=15)
plt.ylabel('Price in USD', fontsize=15)
plt.legend(loc='upper right', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
per_item_aap = df.groupby(['Country', 'Year', 'Food Item'])['Price in USD'].mean().reset_index()

sns.set_style('whitegrid')

food_items = df['Food Item'].unique()

for item in food_items:
    plt.figure(figsize=(15, 8))
    sns.barplot(data=per_item_aap[per_item_aap['Food Item']==item],
                x='Year', y='Price in USD', hue='Country', palette='tab10')
    plt.title('Annual Average Prices of {} in Different Countries'.format(item))
    plt.xlabel('Year', fontsize=16)
    plt.ylabel('Price in USD', fontsize=16)
    
    plt.legend(bbox_to_anchor=(1.05,1), loc=2, borderaxespad=0., fontsize=14)
    
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    
    plt.show()
```

```python
pivot_table = df.pivot_table(values='Price in USD', index='Date', columns=['Country', 'Food Item'])

corr_matrix = pivot_table.corr()

plt.figure(figsize=(15,12))
sns.heatmap(corr_matrix, annot=True, cmap='viridis', fmt='.2f', annot_kws={'size':12})
plt.title('Correlation Heatmap of Prices', fontsize=20)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
#Converting categorical variables (Country and Food names) into numeric inputs using OneHotEncoder
#Scaling using StandardScaler ensures that all numeric variables are on a consistent scale.
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), ['Country', 'Food Item']),
        ('num', StandardScaler(), ['Year', 'Month'])
    ])

df_preprocessed = preprocessor.fit_transform(df)
```

```python
#Splitting the dataset: 80% of the data will be used for training the model and the reamining 20% for testing its performance
from sklearn.model_selection import train_test_split
df_preprocessed_train, df_preprocessed_test, df_train_target, df_test_target = train_test_split(
    df_preprocessed, df['Price in USD'], test_size=0.2, random_state=42)
```

```python
#Model Selection
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import KFold, cross_val_score
```

```python
rf = RandomForestRegressor(n_estimators=200, max_depth=None, min_samples_split=2, min_samples_leaf=1, random_state=42)
```

```python
#Hyperarameter Tuning
from sklearn.model_selection import RandomizedSearchCV

param_grid = {
    'n_estimators': np.arange(100, 1001, 100),
    'max_depth': [None] + list(np.arange(2, 21)),
    'min_samples_split':np.arange(2, 21),
    'min_samples_leaf': np.arange(1, 21),
    'bootstrap': [True, False]
}
```

```python
random_search = RandomizedSearchCV(estimator=rf, param_distributions=param_grid,
                                   n_iter=100, cv=3, n_jobs=-1, verbose=2, random_state=42)
```

```python
random_search.fit(df_preprocessed_train, df_train_target)

print('Best parameters:', random_search.best_params_)
```

```python
best_rf = RandomForestRegressor(
    n_estimators=random_search.best_params_['n_estimators'],
    max_depth=random_search.best_params_['max_depth'],
    min_samples_split=random_search.best_params_['min_samples_split'],
    min_samples_leaf=random_search.best_params_['min_samples_leaf'],
    bootstrap=random_search.best_params_['bootstrap'],
    random_state=42,
)

best_rf.fit(df_preprocessed_train, df_train_target)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
model = Sequential([
    Dense(128, activation='relu', input_shape=(df_preprocessed_train.shape[1],)), 
    Dropout(0.3), 
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='linear')
])

model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
history = model.fit(df_preprocessed_train, df_train_target, validation_data=(df_preprocessed_test, df_test_target), epochs=100, batch_size=32)
```

```python
rf_train_preds = best_rf.predict(df_preprocessed_train)
rf_test_preds = best_rf.predict(df_preprocessed_test)

nn_train_preds = model.predict(df_preprocessed_train).flatten()
nn_test_preds = model.predict(df_preprocessed_test).flatten()
```

```python
rf_train_rmse = np.sqrt(mean_squared_error(df_train_target, rf_train_preds))
rf_test_rmse = np.sqrt(mean_squared_error(df_test_target, rf_test_preds))

nn_train_rmse = np.sqrt(mean_squared_error(df_train_target, nn_train_preds))
nn_test_rmse = np.sqrt(mean_squared_error(df_test_target, nn_test_preds))
```

```python
print('Random Forest Train RMSE:', rf_train_rmse)
print('Random Forest Test RMSE:', rf_test_rmse)
print('Neural Network Train RMSE:', nn_train_rmse)
print('Neural Network Test RMSE:', nn_test_rmse)
```

#### The Random Forest model outperforms the Neural Network on both the training and testing datasets

```python
#Random Sample Predictions
random_indices = np.random.choice(df_preprocessed_test.shape[0], size=10, replace=False)
random_samples = df_preprocessed_test[random_indices]
random_targets = df_test_target.values[random_indices]
```

```python
rf_predicted_prices = best_rf.predict(random_samples)
nn_predicted_prices = model.predict(random_samples).flatten()
```

```python
comparison_df_rf = pd.DataFrame(data={'Actual Prices': random_targets,'RF PRedicted Prices': rf_predicted_prices})
comparison_df_nn = pd.DataFrame(data={'Actual Prices': random_targets,'NN PRedicted Prices': rf_predicted_prices})
```

```python
colors=plt.cm.viridis(np.linspace(0.5, 1, 2))
```

```python
comparison_df_rf.plot(kind='bar',figsize=(10,8), color=colors)
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.title('Random Forest: Actual vs Predicted Prices')
plt.show()
```

```python
comparison_df_nn.plot(kind='bar',figsize=(10,8), color=colors)
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.title('Neural Network: Actual vs Predicted Prices')
plt.show()
```
